```python
import numpy as np
import pandas as pd

works = pd.DataFrame({
    '작가': ['김유정', '김유정', '김유정',
            '현진건', '현진건',
            '이상', '이상',
            '나도향', '나도향',
            '채만식', '이효석'],
    '작품': ['봄봄', '동백꽃', '만무방',
            '운수 좋은 날', 'B사감과 러브레터',
            '날개', '권태',
            '벙어리 삼룡이', '물레방아',
            '레디메이드 인생', '메밀꽃 필 무렵'],
    '연도': [1935, 1936, 1934,
            1924, 1925,
            1936, 1937,
            1925, 1925,
            1934, np.nan], # 이효석 '메밀꽃 필 무렵' 발표연도가 누락됨
    '글자수': [12500, 9800, 13600,
                11200, 7400,
                18300, 14100,
                16800, 14500,
                22000, 11800],
})

authors = pd.DataFrame({
    '작가': ['김유정', '현진건', '이상', '나도향', '채만식', '이효석', '염상섭'],
    '생년': [1908, 1900, 1910, 1902, 1902, 1907, 1897],
    '몰년': [1937, 1943, 1937, 1926, 1950, 1942, 1963],
})
```

```python
def categorize(n: int) -> str:
    if n > 15000: return '긴'
    if n >= 10000: return '보통'
    return '짧음'    
    
works2['분량'] = works2['글자수'].apply(categorize)
```

Q4 (a) 확인질문
포함되지 않는다. how 인자의 의미는 어느 표를 기준으로 살리는지이기 때문에 right 표(authors)에만 있고 left 표(works)에는 없는, 즉 'left 기준 없는 짝'은 말없이 사라진다.

```python
full['집필 당시 나이'] = full['연도'] - full['생년']

full[['작가', '작품', '연도', '생년','집필 당시 나이']]
```

```python
import pandas as pd

grouped = full.groupby('작가')['집필 당시 나이'].mean()
#작가별 집필 평균 나이를 구하여, 오름차순으로 정렬해 출력

grouped.sort_values()
```
